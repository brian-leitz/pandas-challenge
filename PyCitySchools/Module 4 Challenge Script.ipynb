```python
#import pandas as pd

import pandas as pd 


#load file into document 

loading_school_data = "../../PyCitySchools/Resources/schools_complete.csv"
loading_student_data = "../../PyCitySchools/Resources/students_complete.csv"

#read the files into the dataframe

school_data = pd.read_csv(loading_school_data)
student_data = pd.read_csv(loading_student_data)

#Merge the data into a single dataset

merged_dataframes = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

merged_dataframes.head()
```

```python
#Calculate the total number of unique schools

total_unique_schools = merged_dataframes['school_name'].nunique()
total_unique_schools
```

```python
# Calculate the total number of students

#add value of each column to equal the total number of students in each cell

student_count = merged_dataframes['size'].unique().sum()
student_count
```

```python
# Calculate the total budget

total_budget = merged_dataframes['budget'].unique().sum()
total_budget
```

```python
# Calculate the average (mean) math score

math_average = merged_dataframes['math_score'].mean()
math_average
```

```python
# Calculate the average (mean) reading score

reading_average = merged_dataframes['reading_score'].mean()
reading_average
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)

math_count_pass = merged_dataframes[(merged_dataframes["math_score"] >= 70)].count()["student_name"]
math_percentage_pass = math_count_pass / float(student_count) * 100
math_percentage_pass
```

```python
# Calculate the percentage of students who passed reading 
reading_count_pass = merged_dataframes[(merged_dataframes["reading_score"] >= 70)].count()["student_name"]
reading_percentage_pass = reading_count_pass / float(student_count) * 100
reading_percentage_pass
```

```python
# Use the following to calculate the percentage of students that passed math and reading

math_reading_count_passing = merged_dataframes[
    (merged_dataframes["math_score"] >= 70) & (merged_dataframes["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = math_reading_count_passing / float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame

district_metrics_snapshot = {
    "Total Schools": total_unique_schools,
    "Total Students": student_count,
    "Total Budget": total_budget,
    "Average Math Score": math_average,
    "Average Reading Score": reading_average,
    "% Passing Math": math_percentage_pass,
    "% Passing Reading": reading_percentage_pass,
    "% Overall Passing": overall_passing_rate}

district_df = pd.DataFrame(district_metrics_snapshot, index = [0], columns=["Total Schools",
                                                               "Total Budget", 
                                                               "Total Students", 
                                                               "Average Math Score", 
                                                               "Average Reading Score", 
                                                               "% Passing Math", 
                                                               "% Passing Reading",
                                                               "% Overall Passing"
                                                              
                                                              ])

# Formatting

district_df["Total Students"] = district_df["Total Students"].map("{:,}".format)
district_df["Total Budget"] = district_df["Total Budget"].map("${:,.2f}".format)


# Display the DataFrame
district_df
```
